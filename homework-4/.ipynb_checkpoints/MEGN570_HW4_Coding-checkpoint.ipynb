# MEGN 570 HW 4 - Charge Transfer Modeling

We will model charge transfer at an SOFC anode, where O$^{2-}$ in the electrolyte reacts with 2 hydrogen atoms adsorbed on a Ni surface (H$_{\rm Ni(s)}$) to form adsorbed water (H$_2$O$_{\rm Ni(s)}$) and electrons in the Ni (e$^-_{\rm Ni}$).

$$\mathrm{O^{2-}_{elyte(s)} + 2\,H_{Ni(s)} \leftrightarrows \left[\,\right]_{elyte(s)} + H_2O_{Ni(s)} + \left[\,\right]_{Ni(s)}+ 2\,e^-_{Ni}  }$$

where $\left[\,\right]_{\rm elyte}$ is a surface vacancy on the ceramic electrolyte surface, where the oxide formerly resided, and $ \left[\,\right]_{Ni(s)}$ is a vacancy on the Ni surface.  

We are going to plot the charge transfer current density as a function of the double layer potential $\Delta\phi = \phi_{\rm an} - \phi_{\rm elyte}$ for $-1.0 \leq \Delta \phi \leq 0.15\,{\rm V}$.

For this reaction, the standard concentration for the Ni surface is the total site density, $C^\circ_{\rm Ni(s)} = \Gamma_{\rm Ni(s)} = 2.6\times10^{-8}\,\frac{\rm kmol}{m^2}$, and for the electrolyte it is the total electrolyte surface site density: $C_{\rm elyte}^\circ =\Gamma_{\rm YSZ(s)} =  46.05\,\frac{\rm kmol}{\rm m^2}$

The species thermo and forward rate coefficient are provided for you.

## Part 1:  Elementary mass action kinetics.

Calculate the current density as a function of the electric potential, using elementary mass-action kinetics.  The only requirement is that you provide an equation for `i_elementary[i]`, which calculates the current density ($A m^{-2}$) for each electric potential value.  How you get to that value is up to you.

The actual answers are plotted with your equation, to verify.  

Note that we define positive current as positive charge moving _to_ the electrode, and the electric potential as $\Delta \phi = \phi_{\rm an} - \phi_{\rm elyte}$.  As we are sweeping from $\Delta \phi=0.15$ to $\Delta \phi = -1.0$, should the current get more positive or more negative?

Assume the following conditions:

- $\theta_{\rm H_{Ni(s)}} = 0.6$
- $\theta_{\rm H_2O_{Ni(s)}} = 0.2$
- $\theta_{\rm \left[\,\right]_{Ni(s)}} = 0.2$
- $\theta_{\rm O^{2-}_{elyte(s)}} = 0.92$
- $\theta_{\rm \left[\,\right]_{elyte(s)}} = 0.08$
- $T = 800^\circ$ C

The $k_{\rm fwd}$ value given below is in units that will results in a current given in A m$^{-2}$.

### Import modules:

```python
import numpy as np
from matplotlib import pyplot as plt
from math import exp, log
```

### Givens

```python
F = 96485e3    # Faraday's constant, C/kmol equivalent charge
R = 8314.5     # Gas constant, J/kmol-K
T = 800 + 273.15  #Given Temp (K)

k_fwds = 4.16307062e+7 # Chemical forward rate constant (k_forward_star), m^4/kmol^2/s

C_elyte = 46.05    # Total (reference) elyte concentration, kmol/m3
C_Ni_s = 2.6e-08   # Total (reference) concentration of Ni surface sites, kmol/m2

X_H_Ni = 0.6       #really theta_k = C_k/gamma_0
X_H2O_Ni = 0.2
X_Vac_Ni = 0.2
X_Vac_elyte = 0.08
X_Ox_elyte = 0.92

"Species standard-state thermo"
g_H_Ni_o = -7.109209e+07      # standard-state gibbs energy for H adsorbed on Ni surface (J/kmol)
g_H2O_Ni_o = -3.97403035e+08  # standard-state gibbs energy for H2O adsorbed on Ni surface (J/kmol)
g_Vac_Ni_o = 0.0              # standard-state gibbs energy for Ni surface vacancy (J/kmol)
g_Vac_elyte_o = 0.0           # standard-state gibbs energy for electrolyte oxide vacancy (J/kmol)
g_Ox_elyte_o = -2.1392135e+08 # standard-state gibbs energy for electrolyte oxide O2- (J/kmol)

# Validated (i.e. correct :) results.  We will use these to check our answers:
E_validate = np.array([-0.9, -0.86724138, -0.83448276, -0.80172414, -0.76896552, 
                    -0.7362069, -0.70344828, -0.67068966, -0.63793103, -0.60517241, 
                    -0.57241379, -0.53965517, -0.50689655, -0.47413793, -0.44137931, 
                    -0.40862069, -0.37586207, -0.34310345, -0.31034483, -0.27758621, 
                    -0.24482759, -0.21206897, -0.17931034, -0.14655172,  -0.1137931, 
                    -0.08103448, -0.04827586, -0.01551724,  0.01724138,  0.05])
i_validate = np.array([ 131.74413417,   92.44282478,   64.86313049,   45.50798636,   31.9231884,
                    22.38620407,   15.68775772,   10.97849611,    7.66129177,    5.31552803,
                    3.64376548,    2.43402707,    1.53291973,    0.82618438,    0.22420799,
                    -0.34933914,   -0.96718197,   -1.70766206,   -2.66467121,   -3.95955677,
                    -5.75650834,   -8.28337676,  -11.86056524,  -16.94165607,  -24.17092413,
                    -34.46503005,  -49.12925142,  -70.02299007,  -99.79554155, -142.22202185])
```

### Add some calculations to the following code block:

```python
# Loop over these delta phi = phi_anode - phi_elyte values:
beta = 0.5        #assumption, symmetric transfer coefficient
n = -2

#find activity concentrations for each species
C_ac_H_Ni = X_H_Ni * C_Ni_s
C_ac_H2O_Ni= X_H2O_Ni*C_Ni_s
C_ac_Vac_Ni = X_Vac_Ni*C_Ni_s
C_ac_Vac_elyte = X_Vac_elyte*C_elyte
C_ac_Ox_elyte = X_Ox_elyte*C_elyte

C_fwd = C_ac_Ox_elyte**1*C_ac_H_Ni**2                       #product of species C forward, could be every concentration^(0)
C_rev = C_ac_H2O_Ni**1*C_ac_Vac_elyte**1*C_ac_Vac_Ni**1     #product of species C reverse
C_0 = 1                                                     #product of all species at standard state 

del_G0 = (1*g_H2O_Ni_o +1*g_Vac_elyte_o +1*g_Vac_Ni_o)-(1*g_Ox_elyte_o + 2*g_H_Ni_o) #standard change in Gibbs for overall rxn 
k_revs = k_fwds*(1/(exp(-del_G0/(R*T))*C_0))                #find k_rev_start - solution to 1.1, holds because only chemical

delta_phi = np.linspace(-0.9,0.05,100)

#initialize all arrays in for loop
i_elementary = np.zeros_like(delta_phi)
k_fwd = np.zeros_like(delta_phi)
k_rev = np.zeros_like(delta_phi)

for ind, E in enumerate(delta_phi): 
    k_fwd[ind] = k_fwds*(exp((-beta*n*F*E)/(R*T)))
    k_rev[ind] = k_revs*exp(((1-beta)*n*F*E)/(R*T))
    i_elementary[ind] = n*F*(k_fwd[ind]*C_fwd - k_rev[ind]*C_rev)
```

Plot the results:

```python
fig1, ax1 = plt.subplots()
ax1.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k')
ax1.plot(E_validate,i_validate,'ro',linewidth = 1.5)
ax1.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax1.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
```

```python
fig2, ax2 = plt.subplots()
ax2.semilogy(delta_phi,abs(i_elementary),linewidth = 1.5,color = 'k')
ax2.semilogy(E_validate,abs(i_validate),'ro',linewidth = 1.5)
ax2.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax2.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
```

### Discussion:
One can see that the mass action kinetics approximation follows the validation V, i data well and matches the expectation from theory (though on switched axes then what I've seen before). As electric potential increases, current decreases with an inflection point at the point where overpotential is equal to 0 (about -0.43 V when plotting against electric potential). This behavior is seen more clearly in the semilog plot, plotting electric potential against abs(i). The sharp point of the "V" shaped plot is where the overpotential equals 0. 

## ii. Part 2: Butler-Volmer kinetics  

Now use your derivation to calculate $i_\circ$ as a function of your elementary kinetic parameters, and then calculate the currents for the same range of potentials.

Add some code to the block below:

```python
#initialize all arrays in for loop
i_BV = np.zeros_like(delta_phi)
eta = np.zeros_like(delta_phi)

del_G = del_G0 + R*T*log(C_rev/C_fwd)  #del_G of rxn, dropped C_0 because = 1

C_fwd_BV = C_ac_Ox_elyte**((1-beta)*1)*C_ac_H_Ni**((1-beta)*2)                  #product of species C forward, stoich x by (1-beta)
C_rev_BV = C_ac_H2O_Ni**(beta*1)*C_ac_Vac_elyte**(beta*1)*C_ac_Vac_Ni**(beta*1) #product of species C reverse, stoich x by (beta)
i_o = n*F*k_fwds**(1-beta)*k_revs**(beta)*C_fwd_BV*C_rev_BV                     #xchange current density

for ind, E in enumerate(delta_phi):
    eta[ind] = E + del_G/(n*F)                                                           #overpotential relationship to Gibb's energy
    i_BV[ind] = i_o*(exp((-beta*n*F*eta[ind])/(R*T))-exp(((1-beta)*n*F*eta[ind])/(R*T))) #Butler Volmer Equation
    
    
#print(eta)
```

#### Let's plot the BV and elementary currents together:

```python
plt.close('all')
plt.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k');
plt.plot(delta_phi,i_BV,'ro',markerfacecolor='none');
plt.xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)   #changed, it's not eta we are plotting it's delta_phi
plt.ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(['Mass Action','Butler Volmer'],frameon=False)
plt.show()
```

```python
plt.semilogy(delta_phi,abs(i_elementary),linewidth = 1.5,color = 'k')
plt.semilogy(delta_phi,abs(i_BV),'ro',linewidth = 1.5)
ax2.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax2.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
ax2.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax2.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
```

#### Discussion:

Once again, the Butler Volmer approximation matches the mass action results well, showing the expected form. Also completing a semilog plot, the results still match. I attempted to zoom on different regions to see if there were major departures, but still found well matching data. Regarding the assumption of constant $i_\circ$, I think it must be a good enough approximation for Butler-Volmer to be so widely accepted and generally the standard. But I do think that $i_\circ$ really isn't constant because it's based on forward and reverse reaction rates that depend on species concentrations at any given state, so it seems like it should be changing as reactants are consumed and products are made.  

### C. Tafel current

We can re-use the same parameters ($i_\circ$, $\beta$, etc.) from the Butler-Volmer expression to calculate the current using the Tafel approximation:

```python
i_Tafel_a = np.zeros_like(delta_phi)
i_Tafel_c = np.zeros_like(delta_phi)

alpha_a = abs(n*beta)                           #for elem. rxns, TYP alpha_c+ alpha_a = n = -2, but TYP pos coeff so abs()
alpha_c = abs(n*(1-beta))

#As written, rxn releases electrons, so choose i_Tafel_a, or anodic current
for ind, E in enumerate(delta_phi):
    i_Tafel_a[ind] = i_o*exp((alpha_a*F)/(R*T)*eta[ind])
    i_Tafel_c[ind] = -i_o*exp((-alpha_c*F)/(R*T)*eta[ind])
```

```python
plt.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k');
plt.plot(delta_phi,i_BV,'o',markeredgecolor='r',markerfacecolor='none');

plt.plot(delta_phi,i_Tafel_a,'x',markeredgecolor='b',markerfacecolor='none');
plt.plot(delta_phi,i_Tafel_c,'x',markeredgecolor='m',markerfacecolor='none');

plt.xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)       #changed, it's not eta we are plotting it's delta_phi
plt.ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(['Mass Action','Butler Volmer','Tafel-Anodic', 'Tafel-Cathodic'],frameon=False)

plt.show()
```

### Discussion: 


When modeling the Tafel current, you are saying at certain overpotentials the anodic or cathodic current dominates, and can accurately represent the device current by itself. In my first plot I choose to model both Tafel- Anodic and Tafel-Cathodic in comparison to the mass action and Butler-Volmer currents, for the rest of the comparison plots I hone in specifically on the Tafel-Anodic current. In the first plot, you can see that we are accurately representing about half of the current curvature with each of the electrode currents; we represent anodic current where overpotential is positive and we represent the cathodic current where overpotential is negative. For each, we align best at +/- 0.1 V (look at second and third plot to see this more clearly). This makes sense, because +/- 0.1 V is defined by Fuller as a sufficiently large overpotential to use the Tafel approximation. 

The semilog plots for i_Tafel_a correctly align with the "top half" of the "V", which has the positive overpotentials as well. If we plotted i_Tafel_c, we would align with the bottom half.
